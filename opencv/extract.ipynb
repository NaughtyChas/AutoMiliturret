```python
video = cv2.VideoCapture("./tank2.mp4")
num = 0
step = 20
while True:
    ret,frame = video.read()
    if not ret:
        break
    num += 1
    if num % step == 0:
        cv2.imwrite("./images/" + str(num) + ".jpg",frame)
```
